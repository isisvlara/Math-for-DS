-Define the right Riemann sum function  
-Calculate the width of each rectangle  
-Generate right endpoints  
-Calculate the sum of the areas of the rectangles  
-Return sum of the the areas

```python
import numpy as np
```

```python
def right_riemann_sum(f, a, b, n):
    
    dx = (b - a) / n
    x_values = np.linspace(a + dx, b, n)
    Rn = np.sum(f(x_values) * dx) 
    return Rn
```

# Part 2-A:  
-Defining the function f(x)  
-Defining the interval and number of subintervals (n = 4)  
-Defining right endpoints of subintervals  
-Creating x values for plotting  
-Computing the heights of the rectangles evaluated at the right endpoints  
-Plotting the function  
-Plotting the rectangles for the right Riemann sum  
-Adding labels and showing the graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 4 - x**2

a = 0
b = 2
n = 4
dx = (b - a) / n

x_values = np.linspace(a + dx, b, n)

x_plot = np.linspace(a, b, 500)
y_plot = f(x_plot)

rectangle_heights = f(x_values)

plt.plot(x_plot, y_plot, label='$f(x) = 4 - x^2$', color='blue')

for i in range(n):
    plt.bar(x_values[i], rectangle_heights[i], width=dx, align='edge', edgecolor='black', color='skyblue', alpha=0.6)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of $f(x) = 4 - x^2$ with Right Riemann Sum $R_4$')

plt.grid(True)
plt.legend()
plt.show()
```

# Part 2-B:  
-Defining the function  
-Defining the interval and number of subintervals  
-Definining the right endpoints of the subintervals  
-Computing the right Riemann sum (R4)  

```python
def f(x):
    return 4 - x**2

a = 0
b = 2
n = 4
dx = (b - a) / n

x_values = np.linspace(a + dx, b, n)

R4 = np.sum(f(x_values) * dx)

print(f"The Right Riemann sum R4 is: {R4}")
```

# Part 2-C: 
-Defining the function  
-Creating the right riemann sum function  
-Defining intervals  
-Calculating Riemann sums for different numbers of subintervals  
-Special name is Area Under Curve

```python
def f(x):
    return 4 - x**2

def right_riemann_sum(f, a, b, n):
    dx = (b - a) / n  
    x_values = np.linspace(a + dx, b, n) 
    return np.sum(f(x_values) * dx)

a = 0
b = 2

subintervals = [8, 18, 58, 180, 1800]
for n in subintervals:
    Rn = right_riemann_sum(f, a, b, n)
    print(f"Right Riemann sum with {n} subintervals: {Rn:.6f}")
```
