```python
import numpy as np
```

```python
# Define the matrix A
A = np.array([[4, 0, 1],
              [-1, -6, -2],
              [5, 0, 0]])

# Define the vectors
v1 = np.array([1, 2, 3])
v2 = np.array([0, 1, 0])
```

```python
# Part (a) - Check if v1 is an eigenvector of A
Av1 = np.dot(A, v1)
print("A * v1 =", Av1)
```

```python
# Part (b) - Check if v2 is an eigenvector of A
Av2 = np.dot(A, v2)
print("A * v2 =", Av2)
```

```python
# Part (c) - Find all eigenvalues and eigenvectors of A
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues of A:", eigenvalues)
print("Eigenvectors of A:\n", eigenvectors)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv(r"E:\Rachana Bellevue\video_game_data.csv")

# Extract the relevant columns
user_scores = data['user_score']
critic_scores = data['critic_score']

# (a) Scatterplot of User Scores vs. Critic Scores
plt.figure(figsize=(10, 6))
sns.scatterplot(x=user_scores, y=critic_scores)
plt.title('Scatterplot of User_Scores vs. Critic_Scores')
plt.xlabel('User_Scores')
plt.ylabel('Critic_Scores')
plt.grid(True)

# (b) Principal Component Analysis (PCA)
# Standardize the data
X = data[['user_score', 'critic_score']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)
pca.fit(X_scaled)
components = pca.components_

# Plot the principal components on the scatterplot
origin = scaler.mean_
plt.quiver(*origin, *components[0], scale=3, color='r', label='First Principal Component', angles='xy', scale_units='xy')
plt.quiver(*origin, *components[1], scale=3, color='b', label='Second Principal Component', angles='xy', scale_units='xy')

plt.legend()
plt.show()
```
